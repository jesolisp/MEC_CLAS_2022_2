# Unidad III. Dinámica de una partícula.

**Fuerza**. Interacción con un objeto.

**Fuerza de contacto**. Implica contacto físico entre dos objetos.

**Fuerza de campo**. No involucra contacto físico entre dos objetos.

Las únicas fuerzas *fundamentales* que podemos encontrar en la naturaleza son fuerzas de campo:
 1. fuerzas gravitacionales entre objetos
 2. fuerzas electromagnéticas entre cargas eléctricas,
 3. fuerzas fuertes entre partículas subatómicas y 
 4. fuerzas débiles que surgen en ciertos procesos de decaimiento radiactivo.
 
## Primera ley de Newton (ley de la inercia)
>Si un objeto no interactúa con otros objetos, es posible identificar un marco de referencia en el que el objeto tiene aceleración cero.

Este marco de referencia se llama **marco de referencia inercial**.

>Cualquier marco de referencia que se mueve con velocidad constante en relación con un marco inercial es, en sí mismo, un marco inercial.

La Tierra no es un marco inercial debido a su movimiento orbital en torno al Sol y su movimiento rotacional sobre su propio eje. Sin embargo, al involucrar aceleraciones centrípetas se pueden despreciar ya que son pequeñas en comparación con *g*. Por consiguiente, la Tierra representa un marco inercial.

> Un objeto en reposo se mantiene en reposo y un objeto en movimiento continúa en movimiento con una velocidad constante (esto es, con una rapidez constante en una línea recta) al no haber fuerzas externas bajo un marco de referencia inercial.

**Nota**. Cuando ninguna fuerza actúa sobre un objeto, la aceleración del objeto es cero.

Definimos entonces como **inercia** a la resistencia que presenta un objeto por cambiar su velocidad y a la **fuerza** como aquello que causa un cambio en el movimiento de un objeto.

**Masa**. Es la propiedad de un objeto que indicia cuánta resistencia presenta un objeto para cambiar su velocidad. Es independiente del entorno del objeto y del método utilizado para medirla. Recordemos también que es una cantidad *escalar*.

**Nota**. La masa y el peso son dos cantidades diferentes.
   * **Peso**. Es igual a la magnitud de la fuerza gravitacional ejercida sobre el objeto y varía con la posición.
   
## Segunda ley de Newton 

> La aceleración de un objeto es directamente proporcional a la fuerza neta que actúa sobre él e inversamente
proporcional a su masa cuando se ve desde un marco de refencia inercial:

\begin{equation}
 \label{segln}
 \overrightarrow{\bf a} \varpropt \frac{\sum \overrightarrow{\bf F}}{m}.
\end{equation}

De \eqref{segln}, relacionamos la masa, aceleración y fuerza si se elige una constante de proporcionalidad igual a uno, *i.e*:

\begin{equation}
 \label{segln2}
 \sum \overrightarrow{\bf F} = m\cdot \overrightarrow{\bf a},
\end{equation}

donde $\sum \overrightarrow{\bf F}$ denota la fuerza neta (suma vectorial de todas las fuerzas que actúan sobre el objeto).

**Tarea**. ¿Cuándo es válida la Ec. \eqref{segln2}?

La Ec. \eqref{segln2} es una representación vectorial y es equivalente a una forma de componentes:

\begin{equation}
 \sum F_{x} = m\cdot a_{x}, \quad \sum F_{y} = m\cdot a_{y}, \quad \sum F_{z} = m\cdot a_{z}.
\end{equation}

La unidad de fuerza es el **newton**:

\begin{equation}
 1 \text{N} \equiv 1 \text{kg}\cdot\text{m}/\text{s}^{2}.
\end{equation}

Lo anterior nos indica la fuerza que actua sobre un objeto de 1kg de masa y produce una aceleración de 1$\text{kg}\cdot\text{m}/\text{s}^{2}$.

## Fuerza gravitacional y peso
La Tierra ejerce una fuerza sobre todos los objetos. Esta fuerza es llamada **fuerza gravitacional** y se denota como $\overrightarrow{\bf F}_{g}$. Dicha fuerza se dirige hacia el centro de la Tierra y su magnitud se llama **peso** del objeto.

Cuando un objeto en caída libre experimenta una aceleración $\overrightarrow{\bf g}$ que actúa hacia el centro de la Tierra. Aplicando la segunda Ley de Newton 

\begin{equation}
 \overrightarrow{\bf F} = m \overrightarrow{\bf a},
\end{equation}

a un objeto en caída libre de masa $m$, con $\overrightarrow{\bf a} = \overrightarrow{\bf g}$ y $\sum \overrightarrow{\bf F} = \overrightarrow{\bf F}_{g}$, obtenemos:

\begin{equation}
 \overrightarrow{\bf F}_{g} = m \overrightarrow{\bf g}.
\end{equation}

Por lo tanto, el peso de un objeto queda definido como la magnitud de $\overrightarrow{\bf F}_{g}$

\begin{equation}
 {F}_{g} = mg.
\end{equation}

La ecuación anterior cuantifica la fuerza gravitacional sobre un objeto. También se puede aplicar para calcular la magnitud de la fuerza gravitacional para un objeto fijo o para un objeto sobre el que actúan varias fuerzas.

# Tercera Ley de Newton
> Si dos objetos interactúan, la fuerza $\overrightarrow{\bf F}_{12}$ que ejerce el objeto 1 sobre el objeto 2 es igual en magnitud y opuesta en dirección a la fuerza $\overrightarrow{\bf F}_{21}$ que ejerce el objeto 2 sobre el objeto 1:

\begin{equation}
 \overrightarrow{\bf F}_{12} = -\overrightarrow{\bf F}_{21},
\end{equation}

donde $\overrightarrow{\bf F}_{ab}$ significa "la fuerza que se ejerce *por* $a$ *sobre* $b$".

La fuerza que el pie ejerce sobre el balón se llama *fuerza de acción*, mientras que la fuerza del balón sobre el pie se llama *fuerza de reacción*.

> En todos los casos, las fuerzas de acción y reacción actúan sobre objetos diferentes y deben ser del mismo tipo (gravitacional, eléctrica, etcétera).

Los **diagramas de cuerpo libre** es una representación gráfica que aisla aquellas fuerzas sobre el objeto y elimina las otras fuerzas del análisis.

# Aplicaciones de las Leyes de Newton

## Partícula en equilibrio
Se dice que una **partícula está en equilibrio** si la aceleración del objeto (representado como partícula) es cero. En este modelo, la fuerza neta sobre el objeto es cero:

\begin{equation}
 \sum \overrightarrow{\bf F} = 0.
\end{equation}


## Partícula bajo una fuerza neta
Si un objeto experimenta una aceleración, es posible analizar su movimiento con el modelo de partícula bajo una fuerza neta. Para ello suele utilizarse la segunda Ley de Newton.

Aplicando la segunda Ley de Newton en forma de componentes, podemos observar que la única que fuerza que actúa en la dirección $x$ es $\overrightarrow{\bf T}$

\begin{equation}

\end{equation}
